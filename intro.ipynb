## Problema 1. 

### Extraer correo electronico y telefono de los pacientes por mes.

### - Ahorrar recursos : Mis recursos, los recursos de la empresa. 
### - Las capacidades de mi maquina. 

```python
sheet_names = df_excel.sheet_names[0:12]
main_columns = ['PACIENTE', 'CELULAR', 'CORREO ELECTONICO']
set1 = set(main_columns)


for i in sheet_names: 
    df = pd.read_excel(df_excel, i)
    df = df[main_columns].dropna().reset_index(drop = True)
    print("MES: ", i)
    if set1.issubset(df.columns):
        print("My main columns are available!")
    else: 
        print("My main columns arent available!")
    print('\n')

```

```python
month_names = []
for i in sheet_names: 
    month_names.append(i.lower()[0:3])


for i,j in zip(sheet_names,month_names): 
    df = pd.read_excel(df_excel, i)
    df = df[main_columns].dropna().reset_index(drop = True)
    if df.empty:
        print("El mes de {}, no tiene datos!".format(i))
    else:
        df.to_excel('datasets/{}.xlsx'.format(j))

    print("Archivos creados!!")
```
